```python
import matplotlib.pyplot as plt
```

```python
savings = [0,100,150,200,300,500,350,350,500,600,750,1000]
months = ["Jan","Feb","Mar","Apr","May","June","July","Aug","Sep","Oct","Nov","Dec"]
```

```python
plt.plot(months,savings)
plt.xlabel("Month")
plt.ylabel("Savings($)")
plt.title("Saving Account Results")
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(months,savings)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(months,savings)
```
